```python
import pandas as pd
product = pd.read_csv('Originales/Datos looks/product_2.csv', sep=';', index_col="Unnamed: 0")
product_variant = pd.read_csv('Originales/Datos looks/product_variant.csv')
color = pd.read_csv('Originales/Datos looks/color.csv')
size = pd.read_csv('Originales/Datos looks/size.csv')
brand = pd.read_csv('Originales/Datos looks/brand.csv')
prod_feature_value = pd.read_csv('Originales/Datos looks/product_feature_value.csv')
feature_value = pd.read_csv('Originales/Datos looks/feature_value.csv')
prod_feature_value_qualifier = pd.read_csv('Originales/Datos looks/product_feature_value_qualifier.csv')
feature_value_family = pd.read_csv('Originales/Datos looks/feature_value_family.csv')
feature = pd.read_csv('Originales/Datos looks/feature.csv')
```

```python
import pandas as pd

productos = product[['id', 'title', 'family_id', 'brand_id']]
variantes = product_variant[['product_id', 'season', 'color_id', 'size_id']]

colores = color[['id', 'hexadecimal', 'name']]
colores = colores.rename(columns={'id': 'color_id'})

tallas = size[['id', 'lookiero']]
tallas = tallas.rename(columns={'id': 'size_id'})

marca = brand[['id', 'name', 'origin']]
marca = marca.rename(columns={'id': 'brand_id'})


# producto = pd.DataFrame(dfs['product'].iloc[i]).T
vairante_color = variantes.merge(colores, left_on='color_id', right_on='color_id', how='left')


# df_merged5.to_csv('csv_paraAnder.csv')
# variantes_sin_f
s7 = vairante_color[vairante_color['season'] == 7]
s8 = vairante_color[vairante_color['season'] == 8]
s9 = vairante_color[vairante_color['season'] == 9]

print(f'Season 7 {len(s7)}, Season 8 {len(s8)}, Season 9 {len(s9)}')
```

# Unimos la Tabla de Features

```python
# prod_feature_value + prod_feature_value_qualifier

df1 = prod_feature_value.merge(
    prod_feature_value_qualifier,
    left_on='id',
    right_on='product_feature_value_id',
    how='left',
    suffixes=('', '_qual')
)

# df1 + feature_value

df2 = df1.merge(
    feature_value,
    left_on='feature_value_id',
    right_on='id',
    how='left')

# Renombramos
df2 = df2.rename(columns={'id': 'feature_value_uid'})


#df2 + feature_value_family

df3 = df2.merge(
    feature_value_family,
    left_on='feature_value_id',
    right_on='feature_value_id',
    how='left',
    suffixes=('', '_family'))


# df3 + feature

df_final = df3.merge(
    feature,
    left_on='feature_id',
    right_on='id',
    how='left',
    suffixes=('', '_feature'))

df_final = df_final.rename(columns={'name': 'feature_name'})


# Limpiar Repetidas
cols_drop = [
    'product_feature_value_id',
    'feature_value_id_family',
    'id_feature', 'feature_value_id_qual'    
]

df_final = df_final.drop(columns=[c for c in cols_drop if c in df_final.columns])


### Orden recomendable de columnas

final_cols = [
    'product_id',
    'feature_id',
    'feature_name',
    'feature_value_id',
    'value',
    'family_id',
    'feature_value_id_qual',
    'id',  # id del prod_feature_value
]

df_final = df_final[[c for c in final_cols if c in df_final.columns] + [c for c in df_final.columns if c not in final_cols]]

print(df_final.isna().sum())
df_final['product_id'].nunique()
```

# Ahora combinamos ambas Tablas

```python
import pandas as pd

# Solo features que queremos
features_of_interest = ['top_type', 'adventurous', 'weather', 'print', 'style', 'fit', 'application']

df_features_filtered = df_final[df_final['feature_name'].isin(features_of_interest)]

# Pivot
df_features_pivot = df_features_filtered.pivot_table(
    index='product_id',           # pivot por producto
    columns='feature_name',
    values='value',
    aggfunc='first'
).reset_index()

# Merge features con lookiero_dict (productos + variantes)
final_df = variantes_DEF.merge(
    df_features_pivot,
    left_on='id',
    right_on='product_id',
    how='inner'
)

# Ultima limpieza
final_df = final_df[[
    'id','title','family_id','brand','origin','season',
    'color','hexadecimal',
] + features_of_interest]


adventurous_map = {"one": 1, "two": 2, "three": 3, "four": 4, "five": 5}
final_df['adventurous'] = final_df['adventurous'].map(adventurous_map)

print(final_df.isna().sum()) # los missings que quedan son solo en las features
final_df['family_id'].nunique() # sigue habiendo las mismas familias
len(product) ; len(product_variant) ; len(final_df) # me faltan datos

final_df
```
