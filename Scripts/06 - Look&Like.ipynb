```python
import pandas as pd
```

```python
customers = pd.read_csv('../Datos/Originales/customers_data_2.csv', sep=';')
items = pd.read_csv('../Datos/Originales/items_data.csv', sep=',')
look_like = pd.read_csv('../Datos/Originales/look_and_like_data_2.csv', sep=';')

print('customers:', customers.shape)
print('looklike :', look_like.shape)
print('items    :', items.shape)
```

```python
features = [
    "age",
    "is_mother",
    "season",
    "brand",
    "model",
    "style_1",
    "style_2",
    "place",
    "current_price_eur"   #  PRECIO REAL EN EUROS
]

target = "response"

data = df[features + [target]].dropna()

X = data[features]
y = data[target]

X.head()
```

```python
cat_features = X.select_dtypes(include="object").columns.tolist()
num_features = X.select_dtypes(exclude="object").columns.tolist()

cat_features, num_features
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler


preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_features),
        ("num", StandardScaler(), num_features)
    ]
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

pipeline = Pipeline(
    steps=[
        ("preprocessing", preprocessor),
        ("model", LogisticRegression(
            max_iter=2000,
            class_weight="balanced",
            n_jobs=-1
        ))
    ]
)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

pipeline.fit(X_train, y_train)

y_pred = pipeline.predict(X_test)
y_prob = pipeline.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred))
print("AUC ROC:", roc_auc_score(y_test, y_prob))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
preprocessor_rf = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_features),
        ("num", "passthrough", num_features)
    ]
)

rf_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=12,
    min_samples_leaf=50,
    random_state=42,
    n_jobs=-1,
    class_weight="balanced"
)

pipeline_rf = Pipeline(
    steps=[
        ("preprocessing", preprocessor_rf),
        ("model", rf_model)
    ]
)
```

```python
pipeline_rf.fit(X_train, y_train)

y_pred_rf = pipeline_rf.predict(X_test)
y_prob_rf = pipeline_rf.predict_proba(X_test)[:, 1]

print("RANDOM FOREST RESULTS")
print(classification_report(y_test, y_pred_rf))
print("AUC ROC:", roc_auc_score(y_test, y_prob_rf))
```

```python
threshold = 0.4
y_pred_custom = (y_prob > threshold).astype(int)

print(f"RESULTADOS CON THRESHOLD = {threshold}")
print(classification_report(y_test, y_pred_custom))
```

```python
from xgboost import XGBClassifier
```

```python
xgb_model = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="binary:logistic",
    eval_metric="auc",
    random_state=42,
    n_jobs=-1
)

pipeline_xgb = Pipeline(
    steps=[
        ("preprocessing", preprocessor),  # el mismo de Logistic
        ("model", xgb_model)
    ]
)
```

```python
pipeline_xgb.fit(X_train, y_train)

y_pred_xgb = pipeline_xgb.predict(X_test)
y_prob_xgb = pipeline_xgb.predict_proba(X_test)[:, 1]

print("XGBOOST RESULTS")
print(classification_report(y_test, y_pred_xgb))
print("AUC ROC:", roc_auc_score(y_test, y_prob_xgb))
```

```python
threshold = 0.3
y_pred_xgb_tuned = (y_prob_xgb > threshold).astype(int)

print(f"XGBOOST  THRESHOLD {threshold}")
print(classification_report(y_test, y_pred_xgb_tuned))
```

```python
from catboost import CatBoostClassifier
```

```python
# ndices de columnas categricas
cat_features_cb = [
    X.columns.get_loc(col)
    for col in cat_features
]

X_train_cb = X_train.copy()
X_test_cb = X_test.copy()
```

```python
cb_model = CatBoostClassifier(
    iterations=600,
    depth=6,
    learning_rate=0.05,
    loss_function="Logloss",
    eval_metric="AUC",
    random_seed=42,
    verbose=0
)

cb_model.fit(
    X_train_cb,
    y_train,
    cat_features=cat_features_cb
)
```

```python
y_prob_cb = cb_model.predict_proba(X_test_cb)[:, 1]
y_pred_cb = (y_prob_cb > 0.5).astype(int)

print("CATBOOST  THRESHOLD 0.5")
print(classification_report(y_test, y_pred_cb))
print("AUC ROC:", roc_auc_score(y_test, y_prob_cb))
```

```python
threshold = 0.3
y_pred_cb_tuned = (y_prob_cb > threshold).astype(int)

print(f"CATBOOST  THRESHOLD {threshold}")
print(classification_report(y_test, y_pred_cb_tuned))
```
